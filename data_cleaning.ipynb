# Clasificicación de variables:

- Age -> Categorica / Numerica Discreta (igual pruebo las dos)
- Gender -> Binaria
- Ethnicity -> Categórica
- ParentalEducation -> Categórica
- StudyTimeWeekly -> Numérica Continua
- Absences - Numérica Discreta
- Tutoring -> Binaria
- ParentalSupport-> Categórica
- Extracurricular -> Binaria
- Sports -> Binaria
- Music	-> Binaria
- Volunteering -> Binaria

```python
gender_map = {0: 'Male', 1: 'Female'}
X['Gender'] = X['Gender'].map(gender_map).astype('category')
```

```python
X['Age'] = X['Age'].astype('category')
```

```python
ethnicity_map = {0: 'Caucasian', 1: 'African American', 2: 'Asian', 3: 'Other'}
X['Ethnicity'] = X['Ethnicity'].map(ethnicity_map).astype('category')
```

```python
parental_education_map = {0: 'None', 1: 'High School', 2: 'Some College', 3: 'Bachelor\'s', 4: 'Higher'}
X['ParentalEducation'] = X['ParentalEducation'].map(parental_education_map).astype('category')
```

```python
parental_support_map = {0: 'None', 1: 'Low', 2: 'Moderate', 3: 'High', 4: 'Very High'}
X['ParentalSupport'] = X['ParentalSupport'].map(parental_support_map).astype('category')
```
